# A CNN on FashionMNIST dataset

**I created a convolutional neural network and after training it on the famous FashionMNIST dataset, it reached 90 percent accuracy. Though the number is still far from perfect, the mistakes made by the model are justifiable, since the cases are even hard to distinguish for the human eye.**

### This notebook contains:

-  creating a CNN using Tensorflow/Keras

- fitting it on FashionMNIST dataset

- examination of the mistakes made by the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Input, Conv2D, MaxPooling2D
from tensorflow.keras.datasets import fashion_mnist
import tensorflow
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
item_classes     = {0: 'T-shirt/top', 
                       1: 'Trouser', 
                       2: 'Pullover', 
                       3: 'Dress', 
                       4: 'Coat',
                       5: 'Sandal', 
                       6: 'Shirt', 
                       7: 'Sneaker', 
                       8: 'Bag', 
                       9: 'Ankle boot'}
```

```python
(X_train, y_train), (X_test, y_test)  = fashion_mnist.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0
```

```python
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
input_shape = X_train[0].shape
```

```python
# the model consists of 6 convolution/maxpooling layers, followed by
# dense layers.
# In order to avoid reliance on any particular pixel, there are several Dropout
# layers in the model as well.

model = Sequential([
    Input(shape=input_shape),
    
    Conv2D(32, (3,3), strides=2, activation="relu", padding="same"),
    Conv2D(64, (3,3), strides=2, activation="relu", padding="same"),
    MaxPooling2D((2,2), padding="same"),
    Dropout(0.2),
    
    Conv2D(64, (3,3), strides=2, activation="relu", padding="same"),
    Conv2D(128, (3,3), strides=2, activation="relu", padding="same"),
    MaxPooling2D((2,2), padding="same"),
    Dropout(0.3),
    
    Flatten(),
    Dense(128, activation="relu"),
    Dropout(0.3),
    
    Dense(256, activation="relu"),
    Dropout(0.3),
    
    Dense(10, activation="softmax")
])
```

```python
model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
    )
```

```python
r = model.fit(X_train, y_train, epochs=33, batch_size=20, validation_data=(X_test, y_test))
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
fig, (ax1, ax2) = plt.subplots(1, 2)
    

ax1.set_title('Loss')
ax1.plot(r.history["loss"], label="Train Loss")
ax1.plot(r.history["val_loss"], label="Validation Loss")
ax1.legend()

ax2.set_title('Accuracy')
ax2.plot(r.history["accuracy"], label="Train Accuracy")
ax2.plot(r.history["val_accuracy"], label="Validation Accuracy")
ax2.legend()

plt.show()
```

```python
preds = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, preds)
```

```python
cm
```

```python
# a glance on the most repeated mistakes make by the model
mistakes = [(i,j) for i, j in zip(y_test, preds) if i != j]
common_mistakes = Counter(mistakes).most_common(10)
```

```python
for mistake, count in common_mistakes:
    print(f"True label: {item_classes[mistake[0]]}\t, Predicted: {item_classes[mistake[1]]}\t, Count: {count}")
```

```python
false_predictions = np.where(y_test != preds)
```

```python
# with every run, the next cell will display a randomly-selected
# false prediction of the model
```

```python
plt.rcParams["figure.figsize"] = (7, 7)

i = np.random.choice(false_predictions[0])
plt.title(f"True label: {item_classes[y_test[i]]}, Predicted: {item_classes[preds[i]]}")
plt.imshow(X_test[i][:, :, 0], cmap="gray")
plt.show()
```
